```python
vec_1 = np.array([1, 2, 3, 4, 5, 6])
mat_1 = vec_1.reshape(3, 2)
print("Matrix mat_1:\n", mat_1)
```

```python
def reshape_mat(vector, rows, cols):
    if len(vector) != rows * cols:
        raise ValueError("The size of the vector does not match the specified dimensions.")
    return vector.reshape(rows, cols)

reshaped_mat = reshape_mat(vec_1, 3, 2)
print("Reshaped Matrix:\n", reshaped_mat)
```
