```python
class Calculator:
    def summation(self,n1,n2):
        return n1+n2
    def subtraction(self,n1,n2):
        return n1-n2
    def multiplication(self,n1,n2):
        return n1*n2
    def divition(self,n1,n2):
        return n1/n2 

class Calculator2(Calculator):
    def power(self,n1,n2):
        return n1**n2

c1 = Calculator2() 
print(c1.summation(2,5))
```

```python
correct_username = "admin"
correct_password = "1234"
def login():
    print("Store App")
    attempts = 5
    while attempts > 0:
        username = input("Enter username: ")
        password = input("Enter password: ")

        if username == correct_username and password == correct_password:
            print(" Login successful!")
            break
        else:
            attempts -= 1
            print(" Wrong credentials. Attempts left:", attempts)

    if attempts == 0:
        print(" Account locked!")
login()
```

```python
from prettytable import PrettyTable
products = [{"name":"water","price":80.0,"quantity":1500},
            {"name":"soda","price":130.0,"quantity":1500},
            {"name":"chips","price":75.0,"quantity":1500},
            {"name":"bread","price":45.0,"quantity":1500},
            {"name":"eggs","price":200.0,"quantity":1500}
           ]

table = PrettyTable()
table.field_names = ["product name","price($)","quantity"]
for p in products :
    table.add_row([p["name"],f"{p["price"]:.2f}",p["quantity"]])
print(table)

login()

total_bill = 0
while True:
    product_name = input("\nenter product name (or 'exit' to finsh: )").strip().lower()
    if product_name == 'exit':
        break

    product = next((p for p in products if p['name'].lower() == product_name),None)
    if product is None :
        print("product not found. please try again")
        continue
    try:
        quantity_required = float(input(f"How many {product['name']} do you want ?"))
        if quantity_required <=0:
            print("quantity must be greater than zero")
            continue
        if quantity_required > product['quantity']:
            print("insufficient quantity")
            continue

        discount_persentage = 0.05
        price_befor_disc = quantity_required * product['price']
        discount_amount = price_befor_disc * discount_persentage
        final_price = price_befor_disc - discount_amount

        product['quantity'] -= quantity_required
        total_bill += final_price
        print(f"success! price for {product['name']}: ${final_price:.2f} (saved: ${discount_amount:.2f})")
        another = input("something else? (y/n)").lower()
        if another != 'y':
            break
    except ValueError:
        print("invalled input")
print(f"\nTotal amount to pay: ${total_bill:.2f}")
print("thanks")

```
