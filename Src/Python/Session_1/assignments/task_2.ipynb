```python
msg = "###!!@mocleW EPGTQ!!!6789"

chars = []
for c in msg:
    if c.isalpha() or c == " ":
        chars.append(c)

CleanMsg = "".join(chars)   
words = CleanMsg.split()

FirstWord = list(words[0])
FirstWord.reverse()
FirstWord = "".join(FirstWord)

SecondWord = ""
for c in words[1]:
    if c == "E":
        SecondWord += "A"
    else:
        SecondWord += c

result = FirstWord + " " + SecondWord
print(result)
```

```python
msg = "&&&**$gnirtS PLIO!!@1234"

chars = []
for c in msg:
    if c.isalpha() or c == " ":
        chars.append(c)

CleanMsg = "".join(chars)   
words = CleanMsg.split()

FirstWord = list(words[0])
FirstWord.reverse()
FirstWord = "".join(FirstWord)

SecondWord = ""
for c in words[1]:
    if c == "I":
        SecondWord += "E"
    elif c == "O":
        SecondWord += "I"
    else:
        SecondWord += c

result = FirstWord + " " + SecondWord
print(result)
```
