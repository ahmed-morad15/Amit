```python
def sum(num_1, num_2):
    
    return num_1 + num_2

def sub(num_1, num_2):
    
    return num_1 - num_2
def div(num_1, num_2):
    
    return num_1 / num_2
def multply(num_1, num_2):
    
    return num_1 * num_2

def main():
    print("---------calc------")
    print("choose operations")
    print("1-sum")
    print("2-sub")
    print("3-div")
    print("4-mul")
    
    choice = input("Enter the operation")
    num_1 = float(input("Enter first numb"))
    num_2 = float(input("Enter second numb"))
    
    if choice == "1":
        result = sum(num_1, num_2)
    elif choice == "2":
        result = sub(num_1, num_2)
        
    elif choice == "3":
        result = div(num_1, num_2)
        
    elif choice == "4":
        result = multply(num_1, num_2)
        
    else:
        print("invalid choice")
        return
    
    print("Result: ", result)
    
    
if __name__ == "__main__":
    main()
```

# Solution After Editing

```python
def sum(num_1, num_2):
    return num_1 + num_2

def sub(num_1, num_2):
    return num_1 - num_2

def div(num_1, num_2):
    return num_1 / num_2

def multply(num_1, num_2):
    return num_1 * num_2


def main():
    try:
        print("--------- calc ------")
        print("choose operations")
        print("1-sum")
        print("2-sub")
        print("3-div")
        print("4-mul")

        choice = input("Enter the operation: ")

        num_1 = float(input("Enter first number: "))
        num_2 = float(input("Enter second number: "))

        if choice == "1":
            result = sum(num_1, num_2)

        elif choice == "2":
            result = sub(num_1, num_2)

        elif choice == "3":
            result = div(num_1, num_2)   

        elif choice == "4":
            result = multply(num_1, num_2)

        else:
            print("Invalid choice")
            return

        print("Result:", result)

    except ValueError:
        print("Error: Please enter valid numeric values.")

    except ZeroDivisionError:
        print("Error: Division by zero is not allowed.")

    except Exception as e:
        print("Unexpected error:", e)


if __name__ == "__main__":
    main()
```
