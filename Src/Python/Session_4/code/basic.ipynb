```python
def fact (num):

    """
    this function calculate the factorial of any number given by user
    num : number  
    """

    if num == 0:
        return 1
    
    return num * fact(num - 1)
```

```python
def dividors(num1, num2):
    min_num =min(num1, num2)
    common_div = []
    for i in range(1,min_num+1):
        if num1 % i == 0 and num2 % i == 0:
            common_div.append(i)
    return common_div
```

```python
import random

class PasswordGen:
    def __init__(self):
        self.length = 8
        self.char = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%^&*"

    def gen_pass(self):
        password = ""
        num_of_char = len(self.char)

        for _ in range(self.length):
            ran_index = random.randrange(0, num_of_char)
            password += self.char[ran_index]

        return password


if __name__ == "__main__":
    gen = PasswordGen()
    print(gen.gen_pass())
        
```

```python
class WordCounter:
    def __init__(self, words_list):
        self.words_list = words_list

    def count_words(self):
        unique_words = set(self.words_list)
        count = {word: 0 for word in unique_words}
        for word in unique_words:
            for w in self.words_list:
                if w == word:
                    count[word] += 1
        return count

if __name__ == "__main__":

    words = ["Welcome", "Ali", "Hi", "Ali", "No", "Hi", "No", "Ali", "No", "Ali"]
    counter = WordCounter(words)
    counts = counter.count_words()
    print(counts)
```
